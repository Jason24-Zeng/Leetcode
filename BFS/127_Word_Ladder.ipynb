Given two words beginWord and endWord, and a dictionary wordList, return the length of the shortest transformation sequence from beginWord to endWord, such that:

Only one letter can be changed at a time.
Each transformed word must exist in the word list.
Return 0 if there is no such transformation sequence.

 

Example 1:
```python
Input: beginWord = "hit", endWord = "cog", wordList = ["hot","dot","dog","lot","log","cog"]
Output: 5
Explanation: As one shortest transformation is "hit" -> "hot" -> "dot" -> "dog" -> "cog", return its length 5.
```
Example 2:
```python
Input: beginWord = "hit", endWord = "cog", wordList = ["hot","dot","dog","lot","log"]
Output: 0
Explanation: The endWord "cog" is not in wordList, therefore no possible transformation.
``` 

Constraints:

- 1 <= beginWord.length <= 100
- endWord.length == beginWord.length
- 1 <= wordList.length <= 5000
- wordList[i].length == beginWord.length
- beginWord, endWord, and wordList[i] consist of lowercase English letters.
- beginWord != endWord
- All the strings in wordList are unique.

````python
# single directional BFS
from collections import defaultdict
#collections.deque
class Solution:
    def ladderLength(self, beginWord: str, endWord: str, wordList) -> int:
        if not beginWord or not endWord or not wordList or endWord not in wordList:
            return 0
        visited = {}
        queue = collections.deque([(beginWord, 1)]) # word, step
        word_diction = defaultdict(list)
        length = len(beginWord)
        for word in wordList:
            visited[word] = False
            for i in range(length):
                word_diction[word[:i] + '*'+ word[i+1:]].append(word)
        visited[beginWord] = True
        #print(word_diction)
        while queue:
            word, step = queue.popleft()
            for i in range(length):
                for new_word in word_diction[word[:i] + '*' + word[i+1:]]:
                    if new_word == endWord:
                        return step + 1
                    if visited[new_word] == True:
                        continue
                    queue.append((new_word, step+1))
                    visited[new_word] = True
        return 0
            
````

````python
#bidirectional BFS
from collections import defaultdict
#collections.deque
class Solution:
    def ladderLength(self, beginWord: str, endWord: str, wordList) -> int:
        if not beginWord or not endWord or not wordList or endWord not in wordList:
            return 0
        self.length = len(beginWord)
        self.word_diction = defaultdict(list)
        for word in wordList:
            for i in range(self.length):
                self.word_diction[word[:i]+'*'+word[i+1:]].append(word)
        head_queue = collections.deque([(beginWord, 1)])
        tail_queue = collections.deque([(endWord, 1)])
        visited = {beginWord: 1}
        other_visited = {endWord: 1}
        ans = None
        while head_queue and tail_queue:
            ans = self.nextword(head_queue, visited, other_visited)
            if ans:
                return ans
            ans = self.nextword(tail_queue, other_visited, visited)
            if ans:
                return ans
        return 0
    def nextword(self, queue, visited, other_visited):
        word, step = queue.popleft()
        for i in range(self.length):
            inter_word = word[:i] + '*' + word[i+1:]
            for next_word in self.word_diction[inter_word]:
                if next_word in other_visited:
                    return step + other_visited[next_word]
                if next_word not in visited:
                    visited[next_word] = step + 1
                    queue.append(next_word, step + 1)
        return None
        
````
