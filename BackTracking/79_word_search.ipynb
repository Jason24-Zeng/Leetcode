Given a 2D board and a word, find if the word exists in the grid.

The word can be constructed from letters of sequentially adjacent cells, where "adjacent" cells are horizontally or vertically neighboring. The same letter cell may not be used more than once.
Example 1:

```python
Input: board = [["A","B","C","E"],["S","F","C","S"],["A","D","E","E"]], word = "ABCCED"
Output: true
```

````python
class Solution:
    def exist(self, board, word) -> bool:
        height, width = len(board), len(board[0])
        for i in range(height):
            for j in range(width):
                if board[i][j] == word[0]:
                    if backtracking(i, j, word):
                        return True
        return False
            
        def backtrack(i , j, new_word = word):
            if new_word == "":
                return True
            if i < 0 or i == height or j < 0 or j == width or word[0] != board[i][j]:
                return False
            board[i][j] = '$'
            for hdiff, wdiff in [(0,1),(1,0),(-1,0),(0,-1)]:
                if backtracking(i+hdiff, j+wdiff, word[1:]):
                    return True
            board[i][j] = new_word[0]    
            return False
                
            
````
