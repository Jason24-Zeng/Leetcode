Given a binary tree, determine if it is a valid binary search tree (BST).

Assume a BST is defined as follows:

- The left subtree of a node contains only nodes with keys less than the node's key.
- The right subtree of a node contains only nodes with keys greater than the node's key.
- Both the left and right subtrees must also be binary search trees.
 

Example 1:
```
    2
   / \
  1   3

Input: [2,1,3]
Output: true
```
Example 2:
```
    5
   / \
  1   4
     / \
    3   6

Input: [5,1,4,null,null,3,6]
Output: false
```
Explanation: The root node's value is 5 but its right child's value is 4.

````python
# Solution in Leetcode
# Recursion method

# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution:
    def isValidBST(self, root: TreeNode) -> bool:
        def Valid_leaf(node, lower = float('-inf'), upper = float('inf')):
            if not node:
                return True 
            val = node.val
            if val <= lower or val >= upper:
                return False            
            if not Valid_leaf(node.left, lower, val) or not Valid_leaf(node.right, val, upper):                
                return False
            return True
        return Valid_leaf(root)
            
                
````

````python
# Solution in Leetcode
# Iteration method

# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution:
    def isValidBST(self, root: TreeNode) -> bool:
        stack = [(root, float('-inf'), float('inf'))]
        while stack:
            node, lower, upper = stack.pop()
            if not node:
                continue
            val = node.val
            if val <= lower or val >= upper:
                return False
            stack.append((node.left, lower, val))
            stack.append((node.right, val, upper))
            
        return True
        
````

````python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution:
    def isValidBST(self, root: TreeNode) -> bool:
        if not root:
            return True
        value = root.val
        def determine_true_leaf(node, left, preleft=[None, value, value]): 
            #第一个value是上一个node的值，第二个value是上一个拐点前一个node的值
            if left:
                if not node.left:
                    return True
                else:
                    if preleft[0] == None:
                        # 比较该节点与左子节点的大小
                        if node.left.val >= node.val:
                            return False
                        else:
                            return determine_true_leaf(node.left, True, [None, node.val, node.val]) and determine_true_leaf(node.left, False, [True, node.val, node.val])
                        
                    elif preleft[0] == True:
                        if node.left.val >= node.val or node.left.val <= preleft[2]:
                            return False
                        else:
                            return determine_true_leaf(node.left, True, [True, node.val, preleft[2]]) and determine_true_leaf(node.left, False, [True, node.val, node.val])
                    else: 
        # preleft[0] == False
                        if node.left.val >= node.val or node.left.val <= preleft[2]:
                            return False
                        else:
                            return determine_true_leaf(node.left, True, [True, node.val, preleft[2]]) and determine_true_leaf(node.left, False, [True, node.val, node.val])
            else:
                if not node.right:
                    return True
                else:
                    if preleft[0] == None:
                        if node.right.val <= node.val:
                            return False
                        else: 
                            return determine_true_leaf(node.right, True, [False, node.val, node.val]) and determine_true_leaf(node.right, False, [None, node.val, node.val])
                    elif preleft[0] == True:
                        if node.right.val >= preleft[2] or node.right.val <= node.val:
                            return False
                        else:
                            return determine_true_leaf(node.right, True, [False, node.val, node.val]) and determine_true_leaf(node.right, False,[False, node.val, preleft[2]])
                    else:
                        if node.right.val >= preleft[2] or node.right.val <= node.val:
                            return False
                        else:
                            return determine_true_leaf(node.right, True, [False, node.val, node.val]) and determine_true_leaf(node.right, False, [False, node.val, preleft[2]])
                
        return determine_true_leaf(root, True) and determine_true_leaf(root, False)
````
