```

2187 seven digit numbers, 6561 eight digit numbers, and 19683 nine digit numbers.

```

````python
class Solution:
    def findnumTree(self, n: int):
        dp = [1] + [0 for i in range(n)]
        for i in range(1, n+1):
            for j in range(i):
                dp[i] += dp[j] * dp[i-j-1]
        return dp
    
````
